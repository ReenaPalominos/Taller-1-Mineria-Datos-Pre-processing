```python
data6.info()
```

```python
# Se revisa la categoria "TIPO DE LICENCIA"
data6.groupby("TIPO DE LICENCIA").count()
```

```python
# Se revisa la categoria "TIPO LICENCIA SEGÚN CONTRALORÍA"
data6.groupby("TIPO LICENCIA SEGÚN CONTRALORÍA").count()
```

```python
# Se agruparan los datos de Tipo de licencia, con Tipo de licencia segun contraloria, debido a que las 2 poseen las mismas opciones.
# En el caso de que coincidan las 2 opciones, existira solo un "1" por fila, en el caso de que difieran, habran dos valores "1" por fila, representando cada valor

# Creando columna binaria "Accidente del trabajo"

data6["Accidente del Trabajo"] = data6.apply(
    lambda x:
        1 if ((x["TIPO DE LICENCIA"] == "Accidente del Trabajo") |
              (x["TIPO LICENCIA SEGÚN CONTRALORÍA"] == "Accidente del Trabajo")) else 0,
        axis = 1
)

# Creando columna binaria "Enf. Hijo Menor"

data6["Enf. Hijo Menor"] = data6.apply(
    lambda x:
        1 if ((x["TIPO DE LICENCIA"] == "Enf. Hijo Menor")|
              (x["TIPO LICENCIA SEGÚN CONTRALORÍA"] == "Enf. Hijo Menor")) else 0,
        axis = 1
)
# Creando columna binaria "Enf. o Acc. no del Trabajo"

data6["Enf. o Acc. no del Trabajo"] = data6.apply(
    lambda x:
        1 if ((x["TIPO DE LICENCIA"] == "Enf. o Acc. no del Trabajo")|
              (x["TIPO LICENCIA SEGÚN CONTRALORÍA"] == "Enf. o Acc. no del Trabajo")) else 0,
        axis = 1
)

# Creando columna binaria "Enfermedad Profesional"

data6["Enfermedad Profesional"] = data6.apply(
    lambda x:
        1 if ((x["TIPO DE LICENCIA"] == "Enfermedad Profesional")|
              (x["TIPO LICENCIA SEGÚN CONTRALORÍA"] == "Enfermedad Profesional")) else 0,
        axis = 1
)

# Creando columna binaria "Licencia Maternal"

data6["Licencia Maternal"] = data6.apply(
    lambda x:
        1 if ((x["TIPO DE LICENCIA"] == "Licencia Maternal")|
              (x["TIPO LICENCIA SEGÚN CONTRALORÍA"] == "Licencia Maternal")) else 0,
        axis = 1
)

# Creando columna binaria "Patologia del Embarazo"

data6["Patologia del Embarazo"] = data6.apply(
    lambda x:
        1 if ((x["TIPO DE LICENCIA"] == "Patologia del Embarazo")|
              (x["TIPO LICENCIA SEGÚN CONTRALORÍA"] == "Patologia del Embarazo")) else 0,
        axis = 1
)

# Creando columna binaria "Prorroga Medicina Preventiva"

data6["Prorroga Medicina Preventiva"] = data6.apply(
    lambda x:
        1 if ((x["TIPO DE LICENCIA"] == "Prorroga Medicina Preventiva")|
              (x["TIPO LICENCIA SEGÚN CONTRALORÍA"] == "Prorroga Medicina Preventiva")) else 0,
        axis = 1
)
```
